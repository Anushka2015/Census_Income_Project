```python
plt.figure(figsize=(30, 30))
plt.suptitle('Univariate Analysis of categoric Features', fontsize=50, fontweight='bold',
alpha=0.8, y=1.)
for i in range(0, len(categoric_features)):
    plt.subplot(5, 3, i+1)
    sns.countplot(x=df1_copy[categoric_features[i]],color='b')
    plt.xlabel(categoric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold',
alpha=0.8, y=1.)
for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df1_copy[numeric_features[i]],shade=True, color='b')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
